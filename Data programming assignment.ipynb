//Question 3:
public class area
{
    public static void Main(string[] args)
    {
        Console.WriteLine(triangle(4,5,6));
    }

    static double triangle(int num1, int num2, int num3)
    {
        double semi_peri = (num1 + num2 + num3) / 2.0;
        double area_tri = Math.Sqrt(semi_peri * (semi_peri - num1) * (semi_peri - num2) * (semi_peri - num3));
        return area_tri;
    }
}

```python
def bldcount(name):
    myfile = open(f"{name}")
    for item in myfile:
        y = list(item.split(' '))
    
    bloodtypes = ['AB','O','A','B','OO']
    
    for item in bloodtypes:
        if y.count(item) > 1:
            print(f"There are {y.count(item)} patients of blood type {item}.")
        elif y.count(item) == 1:
            print(f"There is one patient of blood type {item}.")
        else:
            print(f"There are no patients of blood type {item}.")
```
